```python

# --- Paths ---
img_path = "/kaggle/working/Tumor-1/test/images/y1075_jpg.rf.8687bb518474cb0244062e1beae56d9e.jpg"
label_path = "/kaggle/working/Tumor-1/test/labels/y1075_jpg.rf.8687bb518474cb0244062e1beae56d9e.txt"

# --- Load model ---
model = YOLO("best-model.pt")

# --- Load image ---
img = cv2.imread(img_path)
h, w, _ = img.shape

# --- Read ground truth boxes ---
gt_boxes = []
if os.path.exists(label_path):
    with open(label_path, "r") as f:
        for line in f.readlines():
            cls, x_center, y_center, bw, bh = map(float, line.strip().split())
            x1 = int((x_center - bw/2) * w)
            y1 = int((y_center - bh/2) * h)
            x2 = int((x_center + bw/2) * w)
            y2 = int((y_center + bh/2) * h)
            gt_boxes.append([x1, y1, x2, y2, int(cls)])

# --- Run prediction ---
results = model.predict(img_path, conf=0.25)
pred_boxes = results[0].boxes.xyxy.cpu().numpy()
pred_labels = results[0].boxes.cls.cpu().numpy()

# --- Prepare images for plotting ---
# Original image
img_orig = cv2.cvtColor(img.copy(), cv2.COLOR_BGR2RGB)

# Image with predicted boxes
img_pred = img.copy()
for i, box in enumerate(pred_boxes):
    x1, y1, x2, y2 = map(int, box)
    cls = int(pred_labels[i])
    cv2.rectangle(img_pred, (x1, y1), (x2, y2), (255, 0, 0), 2)
    cv2.putText(img_pred, f"Pred {cls}", (x1, y1-5),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,0,0), 1)
img_pred = cv2.cvtColor(img_pred, cv2.COLOR_BGR2RGB)

# Image with ground truth boxes
img_gt = img.copy()
for box in gt_boxes:
    x1, y1, x2, y2, cls = box
    cv2.rectangle(img_gt, (x1, y1), (x2, y2), (0, 255, 0), 2)
    cv2.putText(img_gt, f"GT {cls}", (x1, y1-5),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 1)
img_gt = cv2.cvtColor(img_gt, cv2.COLOR_BGR2RGB)

# --- Plot side by side ---
fig, axes = plt.subplots(1, 3, figsize=(18,6))
axes[0].imshow(img_orig)
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(img_pred)
axes[1].set_title("Predicted Boxes")
axes[1].axis('off')

axes[2].imshow(img_gt)
axes[2].set_title("Ground Truth Boxes")
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
